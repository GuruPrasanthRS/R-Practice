# Principal Component Analysis(PCA)

## Load necessary library


```r
library(ggplot2)
```

## Load the iris dataset

```r
data(iris)
```

## Remove the categorical variable (Species) for PCA

```r
iris_numeric <- iris[, 1:4]
```

## Perform PCA

```r
pca_result <- prcomp(iris_numeric, center = TRUE, scale. = TRUE)
```

## Print summary of PCA

```r
summary_pca <- summary(pca_result)
print(summary_pca)
```

## Print PCA loadings

```r
print(pca_result$rotation)
```

## Create a biplot

```r
biplot(pca_result, scale = 0)
```

## Convert PCA results to a dataframe

```r
pca_df <- data.frame(pca_result$x, Species = iris$Species)
```

## Plot the first two principal components

```r
p <- ggplot(pca_df, aes(PC1, PC2, color = Species)) +
  geom_point(size = 3) +
  theme_minimal() +
  ggtitle("PCA of Iris Dataset")

print(p)
```

## Determine the best principal component based on explained variance

```r
explained_variance <- summary_pca$importance[2, ]
best_pc <- which.max(explained_variance)
cat("The best principal component is PC", best_pc, "with", round(explained_variance[best_pc] * 100, 2), "% variance explained.\n")
```
